Importando as bibliotecas necessárias

```python
import pandas as pd
import numpy as np
import tkinter as tk
from tkinter import filedialog
```

Seleção do arquivo com os dados

```python
root = tk.Tk()
root.attributes('-topmost', True) # Garante que a janela apareça na frente
root.withdraw() # Esconde a janelinha preta extra

print("Aguardando você selecionar o arquivo na janela...")
caminho_escolhido = filedialog.askopenfilename(
    title="Selecione o seu arquivo do banco de dados",
    filetypes=[("Arquivos CSV", "*.csv"), ("Todos os arquivos", "*.*")]
)
# Se foi selecionadao um arquivo (e não no botão 'Cancelar')
if caminho_escolhido:
    print(f"\nSucesso! Abrindo o arquivo...\n")
```

Carregando o arquivo

```python
 df = pd.read_csv(caminho_escolhido)
```

Limpeza dos dados

```python
 colunas_com_erro = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
    df[colunas_com_erro] = df[colunas_com_erro].replace(0, np.nan)
    
    # Cria o DataFrame final, jogando fora as linhas com buracos
    df_limpo = df.dropna()
    
    print("--- RELATÓRIO DE LIMPEZA ---")
    print(f"Linhas originais (com erros): {len(df)}")
    print(f"Linhas mantidas (limpas e reais): {len(df_limpo)}")
    print("-" * 28 + "\n")
    
```

Exploração dos dados limpos

```python
    print("--- Primeiras 5 linhas do DataFrame Limpo ---")
    print(df_limpo.head())
    print("\n")

    print("--- Informações Gerais ---")
    df_limpo.info()
    print("\n")

    # Contar o número de dados
    num_linhas, num_colunas = df_limpo.shape
    print("--- Contagem Final de Dados ---")
    print(f"O banco de dados possui {num_linhas} registros e {num_colunas} atributos.\n")
    
```

Estatistica Descritiva

```python
    print("--- Estatística Descritiva ---")
    # O comando describe() calcula a média, desvio padrão, mínimo, máximo etc
    estatisticas = df_limpo.describe()
    print(estatisticas)
    print("\n")
```

Analise Avançada

```python
    print("--- Análise Exploratória: Médias agrupadas por Diagnóstico (Outcome) ---")
    # Agrupa por diagnóstico (0 = Sem Diabetes, 1 = Com Diabetes) e tira a média
    # Com isso comparamos diretamente a saúde dos dois grupos
    medias_por_diagnostico = df_limpo.groupby('Outcome').mean()
    print(medias_por_diagnostico)
    print("\nProcessamento concluído com sucesso!")

else:
    print("\nVocê cancelou a seleção. Nenhum arquivo foi aberto.")
```
